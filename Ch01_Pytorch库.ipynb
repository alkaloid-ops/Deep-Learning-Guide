### 自动微分

```python
x = torch.arange(1.0, 5) #设定输入数据x
y = torch.arange(2.0,9,2) #设定真实数据y
display(x,y)
```

```python
w = torch.tensor(1.0, requires_grad=True) #设置初始权重w（启用梯度）
w       #只对需要优化的参数开启梯度追踪，其他参数无需开启（损失函数和中间激活函数计算将自动继承梯度追踪）
```

```python
yhat = w*x
yhat
```

```python
loss = ((yhat-y)**2).mean()
loss
```

```python
loss.backward()
```

```python
w.grad
```

### 自动微分的梯度下降实现

```python
def gradient(x,y,epoch,lr):
    #随机初始化权重和截距（创建与1行n列的权重向量，一个标量）
    w = torch.randn((1,x.shape[1]), dtype=torch.float32, requires_grad=True)
    b = torch.randn(1, dtype=torch.float32, requires_grad=True)

    for epoch in range(epoch):
        loss = torch.mean((torch.mm(x,w.T)+b-y)**2) #计算MSE损失
        loss.backward()                             #损失求梯度（反向传播）
        
        with torch.no_grad():                       #声明以下操作pytorch无需追踪梯度（阻止计算图追踪）
            w -= lr * w.grad                        #更新权重
            b -= lr * b.grad                        #更新截距

        w.grad.zero_()                              #梯度重置为0
        b.grad.zero_()                              #梯度重置为0
        
        if epoch %50 == 0:
            print(f'迭代{epoch}次: w:{w}; loss:{loss}')

    return w,b,loss
```
